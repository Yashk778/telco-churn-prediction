```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest ,chi2
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
```

```python
df = pd.read_csv('clean.csv')
```

```python
df.drop(columns = , axis=1, inplace=True)
```

```python
df.drop(columns = 'Unnamed: 0', axis=1, inplace=True)
```

```python
df
```

Spliting Data

```python
x = df.drop('Churn' , axis = 1)
y = df['Churn']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

**ENCODING**

```python
cat_cols= [
    'gender','Partner','Dependents','Contract','OnlineBackup','OnlineSecurity',
    'PaymentMethod','InternetService','PhoneService','DeviceProtection',
    'TechSupport','StreamingTV','Churn','MultipleLines','StreamingMovies','PhoneService'
]
for col in cat_cols:
    print(col, ":", df[col].unique())
```

```python
ordinal_cols = ['Contract']
ohe_cols = ['OnlineBackup','OnlineSecurity','PaymentMethod','InternetService','DeviceProtection','TechSupport','StreamingTV','MultipleLines','StreamingMovies']
# selector = SelectKBest(score_function  = chi2,k = 10) 
```

```python
binary_cols = ['Partner','Dependents','PhoneService','Churn','gender']

for col in binary_cols:
    df[col] = df[col].map({'No':0,'Yes':1})
```

```python
encoder = ColumnTransformer([
    ('ord' , OrdinalEncoder() , ordinal_cols),
    ('ohe' , OneHotEncoder() , ohe_cols)
] ,remainder = 'passthrough')
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "XGBoost": XGBClassifier()
}
```

```python
from sklearn.metrics import accuracy_score
from sklearn.pipeline import Pipeline

for name, model in models.items():
    
    pipe = Pipeline([
        ('encoder', encoder),
        ('model', model)])
    
    pipe.fit(x_train, y_train)
    y_pred = pipe.predict(x_test)
    
    acc = accuracy_score(y_test, y_pred)
    print(f"{name} Accuracy: {acc:.4f}")
```
